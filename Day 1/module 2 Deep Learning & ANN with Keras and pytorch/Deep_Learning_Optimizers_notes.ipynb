# what is momentum ? why we take 0.9

Momentum is a hyperparameter used in optimization algorithms like SGD. It helps accelerate SGD in the relevant direction and dampens oscillations. It does this by adding a fraction of the update vector of the past timestep to the current update vector. This helps the optimizer overcome local minima and converge faster.

A momentum value of 0.9 is a commonly used default value because it provides a good balance between accelerating convergence and avoiding overshooting the minimum. It's a hyperparameter that is often tuned during model training to find the optimal value for a specific dataset and model architecture.

```python
from tensorflow.keras.optimizers import SGD
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense

# Replace (20,) with your actual input shape, e.g., (784,) for flattened images
input_shape = (20,)

model = Sequential([
    Dense(64, activation='relu', input_shape=input_shape),
    Dense(1, activation='sigmoid') # Replace 1 and sigmoid with your desired output layer
])

# Define the Stochastic Gradient Descent (SGD) optimizer
optimizer = SGD(learning_rate=0.01, momentum=0.9)
# Compile the model with the specified optimizer and mean squared error (mse) loss
model.compile(optimizer=optimizer, loss='mse')
```

#ðŸ“Œ 3. RMSprop

âœ… Explanation:

>Adapts learning rate for each parameter using exponentially weighted average of past squared gradients.

>Especially useful for RNNs and noisy gradients.

ðŸ“˜ Equation:

image.png

```python
from tensorflow.keras.optimizers import RMSprop

optimizer = RMSprop(learning_rate=0.001)
model.compile(optimizer=optimizer, loss='categorical_crossentropy')
```

#ðŸ“Œ 4. Adam (Adaptive Moment Estimation)

âœ… Explanation:

>Combines momentum and RMSprop.

>Maintains moving averages of both gradients and their squares.

```python
from tensorflow.keras.optimizers import Adam

optimizer = Adam(learning_rate=0.001, beta_1=0.9, beta_2=0.999)
model.compile(optimizer=optimizer, loss='mse')
```

#âœ… Summary Table

| Optimizer | Key Feature                   | Best Use Case                   |
| --------- | ----------------------------- | ------------------------------- |
| SGD       | Simple gradient update        | Small datasets, tuning required |
| Momentum  | Adds momentum to SGD          | Oscillating gradient scenarios  |
| RMSprop   | Adaptive LR, fast convergence | RNNs, noisy data                |
| Adam      | RMSprop + Momentum            | Default, general-purpose choice |
| Adagrad   | Per-parameter adaptive LR     | Sparse data (NLP)               |
| Adadelta  | No manual LR tuning           | Deep networks, online learning  |
| Nadam     | Adam + Nesterov lookahead     | NLP and large vision models     |
